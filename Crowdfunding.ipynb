```python
import tkinter as tk
from tkinter import messagebox
import tkinter.messagebox as messagebox
from datetime import datetime
import re
```

```python
class ModifyForm:
    data_other_user = ""
    id=0

    def __init__(self):
        self.window = tk.Tk()
        self.window.title("Project Manager")
        self.window.geometry("600x500")

        self.text_widget = tk.Text(self.window, wrap="word")
        self.text_widget.grid(row=1, column=1, columnspan=3, pady=(5, 0), sticky="nsew")

        load_button = tk.Button(self.window, text="Load Data", command=self.load_data)
        load_button.grid(row=2, column=1)

        delete_button = tk.Button(self.window, text="Delete Project", command=self.delete_project)
        delete_button.grid(row=2, column=2)

        edit_button = tk.Button(self.window, text="Edit Project", command=self.edit_project)
        edit_button.grid(row=2, column=3)

        # Entry fields for project details
        self.title_entry = tk.Entry(self.window, width=20)
        self.title_entry.grid(row=3, column=2)
        self.details_entry = tk.Entry(self.window, width=20)
        self.details_entry.grid(row=4, column=2)
        self.target_entry = tk.Entry(self.window, width=20)
        self.target_entry.grid(row=5, column=2)
        self.star_entry = tk.Entry(self.window, width=20)
        self.star_entry.grid(row=6, column=2)
        self.end_entry = tk.Entry(self.window, width=20)
        self.end_entry.grid(row=7, column=2)

        # Labels for entry fields
        tk.Label(self.window, text="Title:").grid(row=3, column=1)
        tk.Label(self.window, text="Details:").grid(row=4, column=1)
        tk.Label(self.window, text="Total Target (EGP):").grid(row=5, column=1)
        tk.Label(self.window, text="Start Time:").grid(row=6, column=1)
        tk.Label(self.window, text="End Time:").grid(row=7, column=1)

        # Configure row and column weights for resizing
        self.window.grid_rowconfigure(1, weight=1)
        self.window.grid_columnconfigure(1, weight=1)
        self.window.grid_columnconfigure(2, weight=1)
        self.window.grid_columnconfigure(3, weight=1)
    
    def load_data(self):
        text = ""
        try:
            # Replace 'funding_data.txt' with your actual file path
            file_path = 'funding_data.txt'

            # Read data from the chosen file
            with open(file_path, 'r') as file:
                data = file.readlines()

            # Assuming LoginForm has a variable named new_id
            Login_Form = LoginForm()
            self.id = Login_Form.new_id

            for line in data:
                parts = line.strip().split(',')
                if parts[0] == self.id:
                    text += line + "\n"
                else:
                    ModifyForm.data_other_user += line

            # Display the data
            self.text_widget.delete(1.0, tk.END)
            self.text_widget.insert(tk.END, text)

        except FileNotFoundError:
            tk.messagebox.showerror("Error", "File not found.")

    def delete_project(self):
        project_title = self.title_entry.get()

        if not project_title:
            messagebox.showwarning("Warning", "Please enter a project title.")
            return

        # Search for lines containing the specified project title
        all_text = self.text_widget.get(1.0, tk.END)
        lines_with_title = [line for line in all_text.split('\n') if project_title in line]

        if not lines_with_title:
            messagebox.showwarning("Warning", "Project not found.")
            return

        # Remove the lines with the specified project title from the text widget
        for line in lines_with_title:
            all_text = all_text.replace(line, '')

        # Update the text widget
        self.text_widget.delete(1.0, tk.END)
        self.text_widget.insert(tk.END, all_text)

        # Save the updated content to the file
        self.save_to_file(all_text)

        messagebox.showinfo("Success", "Project deleted successfully.")

    def edit_project(self):
        title = self.title_entry.get()
        details = self.details_entry.get()
        target_amount = self.target_entry.get()
        start_date = self.star_entry.get()
        end_date = self.end_entry.get()

        if not title or not details or not target_amount or not start_date or not end_date:
            messagebox.showerror("Error", "Please fill in all fields.")
            return

        # Search for lines containing the specified project title
        all_text = self.text_widget.get(1.0, tk.END)
        lines_with_data = [line for line in all_text.split('\n') if title in line]

        if not lines_with_data:
            messagebox.showwarning("Warning", "Project not found.")
            return

        # Display the details of the first project with the specified title in the entry fields
        project_details = lines_with_data[0].split(',')
        self.details_entry.delete(0, tk.END)
        self.details_entry.insert(0, project_details[2])

        self.target_entry.delete(0, tk.END)
        self.target_entry.insert(0, project_details[3])

        self.star_entry.delete(0, tk.END)
        self.star_entry.insert(0, project_details[4])

        self.end_entry.delete(0, tk.END)
        self.end_entry.insert(0, project_details[5])

        # Update the details of the first project with the specified title in the content
        project_details = lines_with_data[0].split(',')
        updated_line = f"{self.id},{title},{details},{target_amount},{start_date},{end_date}\n"
        all_text = all_text.replace(lines_with_data[0], updated_line)

        # Update the text widget
        self.text_widget.delete(1.0, tk.END)
        self.text_widget.insert(tk.END, all_text)

        # Save the updated content to the file
        self.save_to_file(all_text)

        messagebox.showinfo("Success", "Project details loaded for editing.")

    def save_to_file(self, content):
        try: 
            with open("funding_data.txt", 'w') as file:
                cleaned_content = (content + ModifyForm.data_other_user).strip()
                file.write(cleaned_content)
                            

        except FileNotFoundError:
            tk.messagebox.showerror("Error", "File not found.")

    def run(self):
        self.window.eval('tk::PlaceWindow %s center' % self.window.winfo_toplevel())
        self.window.mainloop()


```

```python

class DisplayDataFromFile:
    def __init__(self):
        self.window = tk.Tk()
        self.window.title("Display Data from File")
        self.window.geometry("400x300")

        load_button = tk.Button(self.window, text="Load Data", command=self.load_data)
        load_button.grid(row=3, column=1)

        search_button = tk.Button(self.window, text="Search by Date", command=self.search_by_date)
        search_button.grid(row=3, column=2)

        # Create entry and button for date search
        self.date_entry = tk.Entry(self.window, width=10)
        self.date_entry.grid(row=3, column=4)

        
        self.text_widget = tk.Text(self.window, wrap="word")
        self.text_widget.grid(row=4, column=1, columnspan=4, pady=(5, 0), sticky="nsew")

        # Configure row and column weights for resizing
        self.window.grid_rowconfigure(4, weight=1)
        self.window.grid_columnconfigure(1, weight=1)
        self.window.grid_columnconfigure(2, weight=1)
        self.window.grid_columnconfigure(3, weight=1)
        self.window.grid_columnconfigure(4, weight=1)

    def load_data(self):
        # Read data from the chosen file
        with open("funding_data.txt", 'r') as file:
            data = file.read()

        # Display the data in the text widget
        self.text_widget.delete(1.0, tk.END)  # Clear existing content
        self.text_widget.insert(tk.END, data)

    def is_valid_date(self, date_string):
        try:
            # Attempt to parse the date
            datetime.strptime(date_string, "%Y-%m-%d")
            return True
        except ValueError:
            return False

    def search_by_date(self):
        search_date = self.date_entry.get()

        if not self.is_valid_date(search_date):
            tk.messagebox.showerror("Error", "Invalid date format. Please use the format YYYY-MM-DD.")
            return

        search_date = datetime.strptime(search_date, "%Y-%m-%d").date()

        # Get all the text from the text widget
        all_text = self.text_widget.get(1.0, tk.END)

        # Clear existing content in the text widget
        self.text_widget.delete(1.0, tk.END)

        # Search for lines containing the specified date
        lines_with_date = [line for line in all_text.split('\n') if search_date.strftime("%Y-%m-%d") in line]

        # Display the matching lines in the text widget
        self.text_widget.insert(tk.END, '\n'.join(lines_with_date))

    def run(self):
        # Center the window on the screen
        self.window.eval('tk::PlaceWindow %s center' % self.window.winfo_toplevel())
        self.window.mainloop()


```

```python

class CreateFund:

    def __init__(self):
        self.window = tk.Tk()
        self.window.title("Home Window")
        self.window.geometry("400x300")

        self.Title_label = tk.Label(self.window, text="Title:")
        self.Title_label.pack()
        self.Title_entry = tk.Entry(self.window)
        self.Title_entry.pack()

        tk.Label(self.window, text="Details:").pack()
        self.details_entry = tk.Entry(self.window)
        self.details_entry.pack()

        tk.Label(self.window, text="Target Amount (EGP):").pack()
        self.target_entry = tk.Entry(self.window)
        self.target_entry.pack()

        tk.Label(self.window, text="Start Date:").pack()
        self.start_entry = tk.Entry(self.window)
        self.start_entry.pack()

        tk.Label(self.window, text="End Date:").pack()
        self.end_entry = tk.Entry(self.window)
        self.end_entry.pack()

        # Create and place the submit button
        submit_button = tk.Button(self.window, text="Submit Campaign", command=self.submit_campaign)
        submit_button.pack()

    def is_valid_date(self, date_string):
        try:
            # Attempt to parse the date
            datetime.strptime(date_string, "%Y-%m-%d")
            return True
        except ValueError:
            return False

    def validate_date(self, start_date, end_date):
        if not self.is_valid_date(start_date) or not self.is_valid_date(end_date):
            return False

        # Convert strings to datetime objects
        start_datetime = datetime.strptime(start_date, "%Y-%m-%d")
        end_datetime = datetime.strptime(end_date, "%Y-%m-%d")

        # Check if start date is before or equal to end date
        return start_datetime <= end_datetime
    
    def submit_campaign(self):
        target=self.Title_entry.get()
        details = self.details_entry.get()
        target_amount = self.target_entry.get()
        start_date = self.start_entry.get()
        end_date = self.end_entry.get()

        if not target or not   details or not target_amount or not start_date or not end_date:
            messagebox.showerror("Error", "Please fill in all fields.")
            return

        if not self.validate_date(start_date, end_date):
            messagebox.showerror("Error", "Invalid date range.")
            return

        #  saving to a file
        Login_Form = LoginForm()
        id =Login_Form.new_id
        
        new_user_data = f"{id},{target},{details},{target_amount},{start_date},{end_date}\n"
        with open("funding_data.txt",'r+') as file:
            lines =file.readlines()
     
            file.write(new_user_data)
            lines.append(new_user_data)
            messagebox.showinfo("Success", "Campaign submitted successfully!")

    def run(self):
        # Center the window on the screen
        self.window.eval('tk::PlaceWindow %s center' % self.window.winfo_toplevel())
        self.window.mainloop()

```

```python
class LoginForm:

    def __init__(self):
        
        self.window = tk.Tk()
        self.window.title("Login Window")
        self.window.geometry("400x300")

        
        center_column = 3
        login_label = tk.Label(self.window, text="Log in", font=("Times New Roman", 25, "bold", "underline", "italic"))
        login_label.grid(row=0, column=center_column, columnspan=2)  

        

        

        # Add Email label 
        self.email_label = tk.Label(self.window, text="Email:")
        self.email_label.grid(row=3, column=center_column)
        self.email_entry = tk.Entry(self.window)
        self.email_entry.grid(row=3, column=center_column + 1)

        # Add Password label 
        self.password_label = tk.Label(self.window, text="Password:")
        self.password_label.grid(row=4, column=center_column)
        self.password_entry = tk.Entry(self.window, show="*")
        self.password_entry.grid(row=4, column=center_column + 1)

        # Add log in button
        self.submit_button = tk.Button(self.window, text="Log in" , command=self.To_home)
        self.submit_button.grid(row=5, column=center_column, columnspan=2)  
    new_id=0
    def To_home(self):
        email = self.email_entry.get()
        password = self.password_entry.get()
        if not  email or not password :
            messagebox.showerror("Error", "Please fill in all fields.")
            return

            # Lists to store all emails and passwords
        all_emails = []
        all_passwords = []

        with open("user_data.txt", 'r') as file_content:
            lines = file_content.readlines()
        for line in lines:
            parts = line.split(',')
            stored_email = parts[3].strip()  
            stored_password = parts[4].strip()
            all_emails.append(stored_email)
            all_passwords.append(stored_password)


        accounts_dict = dict(zip(all_emails,all_passwords))
    
        if email in accounts_dict and password == accounts_dict[email]:
            for line in lines:
                parts = line.split(',')
                if parts[3].strip() == email and parts[4].strip() == password:
                    LoginForm.new_id=parts[0]
                    
            
            Home_form = HomeApp()
            Home_form.run()

        else:
            messagebox.showerror("Error", "Email Or password Not correct.")
            return
    
    def run(self):
        # Center the window on the screen
        self.window.eval('tk::PlaceWindow %s center' % self.window.winfo_toplevel())
        self.window.mainloop()



```

```python
class RegistrationForm:
    
    def __init__(self):        
        self.window = tk.Tk()
        self.window.title("Registration Window")
        self.window.geometry("400x300")
        
        center_column = 3
        login_label = tk.Label(self.window, text="Registration", font=("Times New Roman", 25, "bold", "underline", "italic"))
        login_label.grid(row=0, column=center_column, columnspan=2)  

        # Add First name 
        self.first_name_label = tk.Label(self.window, text="First Name:")
        self.first_name_label.grid(row=1, column=center_column)
        self.first_name_entry = tk.Entry(self.window)
        self.first_name_entry.grid(row=1, column=center_column + 1)

        # Add Last name label 
        self.last_name_label = tk.Label(self.window, text="Last Name:")
        self.last_name_label.grid(row=2, column=center_column)
        self. last_name_entry = tk.Entry(self.window)
        self.last_name_entry.grid(row=2, column=center_column + 1)

        # Add Email label 
        self.email_label = tk.Label(self.window, text="Email:")
        self.email_label.grid(row=3, column=center_column)
        self.email_entry = tk.Entry(self.window)
        self.email_entry.grid(row=3, column=center_column + 1)

        # Add Password label 
        self.password_label = tk.Label(self.window, text="Password:")
        self.password_label.grid(row=4, column=center_column)
        self.password_entry = tk.Entry(self.window, show="*")
        self.password_entry.grid(row=4, column=center_column + 1)

        # Add Submit button
        self.submit_button = tk.Button(self.window, text="Submit" ,command =self.To_home)
        self.submit_button.grid(row=5, column=center_column, columnspan=2)  
        
    

    def validate_email(self,email):
        # Define a regular expression pattern for a basic email validation
        pattern = r'^\S+@\S+\.\S+$'
    
        # Use re.match to check if the email matches the pattern
        if re.match(pattern, email):
            return True
        else:
            return False


    new_id = 0
    def To_home(self):
        first_name = self.first_name_entry.get()
        last_name = self.last_name_entry.get()
        email = self.email_entry.get()
        password = self.password_entry.get()
        if not first_name or not last_name or not email or not password :
            messagebox.showerror("Error", "Please fill in all fields.")
            return

        if not self.validate_email(email):
            messagebox.showerror("Error", "Invalid email .")
            return


        # Read existing data and find the new ID
        with open("user_data.txt", 'r+') as file_content:
            lines = file_content.readlines()
        
            if not lines: 
                RegistrationForm.new_id = 1
            else:
                RegistrationForm.new_id = max(int(line.split(',')[0]) for line in lines) + 1
        
        
        with open("user_data.txt", 'a') as file:
            email_exist = any(email in line for line in lines)
    
            if not email_exist:
              
                
                new_user_data = f"{RegistrationForm.new_id},{first_name},{last_name},{email},{password}\n"
                file.write(new_user_data)
        
                messagebox.showinfo("Success", "User registered successfully!")
            else:
                messagebox.showerror("Warning", "Account with this email already exists.")
                return

            
 
        Login_form = LoginForm()
        Login_form.run() 
        
    def run(self):
        # Center the window on the screen
        self.window.eval('tk::PlaceWindow %s center' % self.window.winfo_toplevel())
        self.window.mainloop()



```
